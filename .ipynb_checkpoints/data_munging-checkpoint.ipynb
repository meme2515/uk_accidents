```python
import pandas as pd
```

```python
acc = pd.read_csv('Acc.csv')
acc.head()
```

```python
acc.columns.tolist()
```

```python
maintain = ['Longitude',
            'Latitude',
            'Accident_Index',
            'Police_Force',
            'Accident_Severity',
            'Number_of_Vehicles',
            'Number_of_Casualties',
            'Date',
            'Day_of_Week',
            'Time',
            'Local_Authority_(District)',
            'Speed_limit',
        ]
acc.drop([c for c in acc.columns if c not in maintain], axis = 1, inplace = True)
acc.head()
```

```python
import xlrd
from copy import copy
wb = xlrd.open_workbook('Road-Accident-Safety-Data-Guide.xls')
wb.sheet_names()
```

```python
def merge_on_guide(df, exclude=[]):
    df2 = df.copy() #Leave the original dataframe unaffected.
    columns = [col for col in df.columns if col not in exclude]
    for sheet in wb.sheet_names():
        sheet_match = sheet.replace(' ', '_')
        if sheet_match in columns:
            print(sheet_match)
            
            replacement = pd.read_excel('Road-Accident-Safety-Data-Guide.xls', sheetname=sheet)\
            .rename(columns={'label' : sheet_match})
            
            df2.rename(columns={sheet_match : '_{}'.format(sheet_match)}, 
                       inplace = True)
            
            df2 = pd.merge(df2, replacement, left_on='_{}'.format(sheet_match), right_on='code').\
            drop(['_{}'.format(sheet_match), 'code'], axis=1)
    return df2
```

```python
acc2 = merge_on_guide(acc)
```

```python
acc2.head()
```

```python
cas = pd.read_csv('Cas.csv')
cas.head()
```

```python
cas.columns.tolist()
```

```python
maintain = ['Accident_Index',
            'Sex_of_Casualty',
            'Car_Passenger',
            'Pedestrian_Road_Maintenance_Worker',
            'Casualty_Type',
            'Age_of_Casualty'
        ]
cas.drop([c for c in cas.columns if c not in maintain], axis = 1, inplace = True)
cas.head()
```

```python
cas2 = merge_on_guide(cas, exclude=['Age_of_Casualty'])
cas2.head()
```
